**Problem Statement:** 
Use a least-square linear regression method to establish a linear function that best fits the data in the table below

| $x$ | 1,000 | 1,500 | 2,000 | 2,500 | 3,000 | 3,500 |
|-----|-------|-------|-------|-------|-------|-------|
| $y$ | 1,457 | 1,282 | 932   | 788   | 465   | 264   |

**Derivation of Solution:** (The derivation of solution is included in the section below)

**Code Listing and Tabulated Solution:** 

Preamble:

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (6, 5)
plt.rcParams['figure.dpi'] = 100
%config InlineBackend.figure_format = 'retina'
```

We first define the data table,

```python
x = np.asarray([1000, 1500, 2000, 2500, 3000, 3500], dtype=np.float64)
y = np.asarray([1457, 1282, 932, 788, 465, 264], dtype=np.float64)
```

We can now determine $a_0$ using Equation 5.8a,

```python
n = len(x)

a0 = (
    (((np.sum(y)) * (np.sum(np.square(x)))) - ((np.sum(np.multiply(x, y))) * (np.sum(x)))) /
    ((n * (np.sum(np.square(x)))) - (np.sum(x) ** 2))
)

a1 = (
    ((n * (np.sum(np.multiply(x, y)))) - (np.sum(x) * np.sum(y))) /
    ((n * (np.sum(np.square(x)))) - (np.sum(x) ** 2))
)

display(a0)
display(a1)
```

```python
g = lambda x: a0 + a1 * x

plt.plot(x, y, 'o')

dx = 0.1
x_range = np.arange(x[0], x[-1] + dx, dx)
plt.plot(x_range, g(x_range))

plt.show()
```
